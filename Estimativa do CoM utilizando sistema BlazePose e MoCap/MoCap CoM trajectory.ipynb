# Série temporal da trajetória do CoM do corpo todo utilizando dados de Mocap coletados e devidamente separados para o trecho desejado.

--> Bibliotecas necessárias

```python
import cv2
import numpy as np
from numpy import asarray
from numpy import save
import pandas as pd
import seaborn; seaborn.set()
```

--> Carregue os dados do código anterior

```python
array = np.load('MoCap CoM gait.npy')
```

--> Salvar os dados referentes à coordenada vertical Y na marcha

```python
import matplotlib.pyplot as plt

data_gait_X = array[:,0]
data_gait_Y = array[:, 1]

np.save('Comy mocap', data_gait_Y)
```

--> Plote o gráfico de Excursão vertical do CoM na marcha

--> Repare que as coordenadas são dadas em milímetros e correspondem às situações do mundo real.

```python
# Definindo a frequência de amostragem
freq_amostragem = 120

t_gait = np.linspace(0, len(data_gait_Y) / freq_amostragem, len(data_gait_Y))

plt.figure(figsize=[15,8])
plt.plot(t_gait, data_gait_Y, color = 'b', linewidth = 5)
plt.ylabel('Deslocamento vertical em mm', fontsize=16)
plt.xlabel('Tempo (s)', fontsize=16)
plt.title('Excursão vertical do CoM na marcha', fontsize=16)
plt.show()
```
